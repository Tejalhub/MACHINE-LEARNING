```python
x=df3.drop('Economic_Impact_Million_USD',axis=1)
y=df3["Economic_Impact_Million_USD"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
LR.fit(xtrain,ytrain)
```

```python
ypred=LR.predict(xtest)
```

```python
ypred
```

```python
TraIn_score=round(LR.score(xtrain,ytrain),2)*100
Test_score=round(LR.score(xtrain,ytrain),2)*100

print(f"Training Result of Linear Regression: {TraIn_score}%")
print(f"Testing Result of Linear Regression: {Test_score}%")
```

```python
from sklearn.tree import DecisionTreeRegressor

#Train
DT= DecisionTreeRegressor(max_depth=4,min_samples_split=2,min_samples_leaf=2)
DT.fit(xtrain,ytrain)

#Predict
y_predict_DT = DT.predict(xtest)
TraIn_score=round(DT.score(xtrain,ytrain),2)*100
Test_score=round(DT.score(xtest,ytest),2)*100

print(f"Training Result of DecisionTreeRegressor: {TraIn_score}%")
print(f"Testing Result of DecisionTreeRegressor: {Test_score}%")
```

```python
from sklearn.neighbors import KNeighborsRegressor

#Train KNN
KNN = KNeighborsRegressor(n_neighbors=25, weights='distance',algorithm='kd_tree',metric='manhattan')
KNN.fit(xtrain, ytrain)

#predict
y_pred_KNN = KNN.predict(xtest)

# Evalute
Training_Acc= round(KNN.score(xtrain,ytrain),2)*100
Testing_Acc= round(KNN.score(xtest,ytest),2)*100
print(f"Training Result of KNeighborsRegressor: {Training_Acc}%")
print(f"Testing Result of KNeighborsRegressor: {Testing_Acc}%")
```

```python
from sklearn.ensemble import RandomForestRegressor

#Train Random Forest
RF = RandomForestRegressor(n_estimators=100, random_state=42)
RF.fit(xtrain, ytrain)

#predict
y_pred_RF = RF.predict(xtest)

# Evalute
Training_Acc= round(RF.score(xtrain,ytrain),2)*100
Testing_Acc= round(RF.score(xtest,ytest),2)*100
print(f"Training Result of RandomForestRegressor: {Training_Acc}%")
print(f"Testing Result of RandomForestRegressor: {Testing_Acc}%")
```

```python
from sklearn.ensemble import GradientBoostingRegressor

GB = GradientBoostingRegressor()
GB.fit(xtrain, ytrain)
y_pred_gbr = GB.predict(xtest)

Training_Acc= round(GB.score(xtrain,ytrain),2)*100
Testing_Acc= round(GB.score(xtest,ytest),2)*100
print(f"Training Result of GradientBoostingRegressor: {Training_Acc}%")
print(f"Testing Result of GradientBoostingRegressor: {Testing_Acc}%")
```

```python
import matplotlib.pyplot as plt

#your model accuracy dictionary
results = {
    'Linear Regression': round(LR.score(xtest, ytest), 2) * 100,
    'Decision Tree': round(DT.score(xtest, ytest), 2) * 100,
    'Random Forest': round(RF.score(xtest, ytest), 2) * 100,
    'KNN': round(KNN.score(xtest, ytest), 2) * 100,
    'GradientBoosting': round(GB.score(xtest, ytest), 2) * 100,
    'ExtremeGradient': round(xgb.score(xtest, ytest), 2) * 100
}

#sort the results by accuracy descending
sorted_results = dict(sorted(results.items(), key=lambda item: item[1], reverse=True))

# Print sorted accuracies
print("\n Final Accuracy Comparison (High to Low):")
print("----------------------------------------")
for model, acc in sorted_results.items():
    print(f"{model}: {acc:.2f}%")
print("-------------------------------------")


#Horizontal Bar Chart (sorted)
plt.figure(figsize=(10,5))
plt.barh(list(sorted_results.keys()), list(sorted_results.values()), color='red')
plt.xlabel('Accuracy (%)' )
plt.title('Model Accuracy Comparison (High to Low)')
plt.gca().invert_yaxis() #To display highest accuracy at the top
plt.tight_layout()
plt.show()
```
